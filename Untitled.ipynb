## Cleaning Data

### Checking for empty values and dealing with them:

```python
df.isna().sum()
```

**Observation:** 
* ph has 491 empty values
* Sulfate has 781 empty values
* Trihalomethanes has 162 empty values

**Action:**
* Remove all rows containing empty values
    (3276 rows × 10 columns) ==>  (2011 rows × 10 columns)
* Or we can fill this values with medians

### 1st Method remove all rows containing empty values

```python
df.dropna(inplace = True)   #remove rows with empty values
#df = df.dropna()           
df.isna().sum()
```

**Observation:**
* The droped rows are more than 1000 and this number is very large where it can affect the model when built 

### 2nd Method refill with means

```python
df[df['Potability'] == 1][['ph', 'Sulfate', 'Trihalomethanes']].mean()
```

```python
df[df['Potability'] == 0][['ph', 'Sulfate', 'Trihalomethanes']].mean()
```

```python
df['ph'].fillna(df['ph'].mean(), inplace= True)
```

```python
df['Sulfate'].fillna(df['Sulfate'].mean(), inplace= True)
```

```python
df['Trihalomethanes'].fillna(df['Trihalomethanes'].mean(), inplace= True)
df.isna().sum()
```

## Modelling

**First data is splitted to be suitable for the model**

```python
from sklearn.model_selection import train_test_split    # Splits The data into four parts for training and testing
```

```python
input_data = df.drop(['Potability'], axis = 1)          # axis = 0 cuts rows - axis = 1 cuts col
target_data = df['Potability']
```

```python
input_train, input_test, target_train, target_test = train_test_split(input_data, target_data,test_size = .2,
                                                                      random_state =42)
```

**After preparing the date the suitable model is imported and built**

### 1] Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(input_train, target_train)     #Train the model to gain experiance
```

```python
pred = model.predict(input_test)        #Test the model
pred
```

```python
np.array(target_test)
```

```python
model.score(input_test, target_test)    #get model accuracy
```

```python
model.coef_          #weights
```

### 2] SVM (Support Vector Machine) Model

```python
from sklearn.svm import SVC
```

```python
svc_model = SVC()
```

```python
svc_model.fit(input_train, target_train)
```

```python
svc_model.score(input_test, target_test)
```

### 3] Tree Model

```python
from sklearn import tree
```

```python
dtree_model = tree.DecisionTreeClassifier()
```

```python
dtree_model.fit(input_train, target_train)
```

```python
dtree_model.score(input_test, target_test)
```

### 4] Random Forest Model

```python
from sklearn .ensemble import RandomForestClassifier
```

```python
random_forest_model = RandomForestClassifier()
```

```python
random_forest_model.fit(input_train, target_train)
```

```python
random_forest_model.score(input_test, target_test)
```

**Observation:** 
* best accuracy: 68.9% by Random Forest ( for filled empty values by mean )
* The accuracy still respectively low to say that the model has high accuracy
